Q1  Create and print 1D, 2D and 3D numPy array:

 a. Print data type of array elements

 b. Print the array shape

 c. Print the dimensions of all arrays

 d. Print the size of all arrays

 e. Transpose of 2D array

```python

import numpy as np
arr_1D = np.array([1, 2, 3, 4, 5])
arr_2D = np.array([[1, 2, 3], [4, 5, 6]])
arr_3D = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
def print_array_details(arr, name):
    print(f"{name}:")
    print(f"Array:\n{arr}")
    print(f"Data Type: {arr.dtype}")
    print(f"Shape: {arr.shape}")
    print(f"Dimensions: {arr.ndim}")
    print(f"Size: {arr.size}\n")
print_array_details(arr_1D, "1D Array")
print_array_details(arr_2D, "2D Array")
print_array_details(arr_3D, "3D Array")
transpose_2D = arr_2D.T
print("Transpose of 2D Array:",transpose_2D)
```

 a. Create two arrays of 3 elements and perform multiplication.

 c. Initialize array of resolution 4 x
 3, initialize a scalar with constant value,  display the shape of 2-D array, output
the 2-D array,  print output of broadcasting scalar value to the array : array_2D + scalar

```python

import numpy as np
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
result_multiplication = arr1 * arr2
print("Array 1:\n",arr1)
print("Array 2:\n",arr2)
print("Result of multiplication:\n",result_multiplication)
print("\n")
arr = np.array([2, 4, 6])
scalar = 3
result_broadcast_scalar = arr * scalar
print("Array:\n",arr)
print("Scalar value:\n",scalar)
print("Result of broadcasting array * scalar:\n",result_broadcast_scalar)
print("\n")
arr_2D = np.arange(12).reshape(4, 3)
scalar_value = 10
print("2D Array (4x3):\n",arr_2D)
print("\nShape of 2D array:\n",arr_2D.shape)
result_broadcast_2D = arr_2D + scalar_value
print("\nScalar value:\n",scalar_value)
print("\nResult of broadcasting scalar + 2D array:\n",result_broadcast_2D)
```

Programs based on Pandas library (data analysis and manipulation)

a. Import a csv file and Print different attributes of dataframe including:
dimension. Shape, column names, data types, print total records using
.head()

b. Implement Access, Add and Drop Operations

c. Demonstrate DataFrame Indexing using .loc() and .iloc()

d. Data Filteration

e. Groupby in Pandas

```python
import pandas as pd

# Load CSV into DataFrame (Assuming a file called 'data.csv')
df = pd.read_csv('/content/iphone.csv')

# Print DataFrame attributes
print("Dimensions (Rows, Columns):", df.shape)
print("Column Names:", df.columns)
print("Data Types:", df.dtypes)
print("\nTotal Records (first 5 rows):\n", df.head())  # Prints first 5 records
```

```python
# Accessing a single column (e.g., 'reviewTitle')
print(df['reviewTitle'])

# Accessing multiple columns (e.g., 'country' and 'ratingScore')
print(df[['country', 'ratingScore']])

# Adding a new column based on existing data
# Example: Creating a new column 'ratingCategory' based on the rating score
df['ratingCategory'] = df['ratingScore'].apply(lambda x: 'High' if x >= 4 else 'Low')

print(df[['ratingScore', 'ratingCategory']].head())  # Display updated DataFrame with the new column

# Dropping a column (e.g., 'reviewUrl')
df_dropped = df.drop('reviewUrl', axis=1)
print(df_dropped.head())  # Display DataFrame after dropping the 'reviewUrl' column

# Dropping rows with a condition (e.g., drop rows where 'ratingScore' is less than 3)
df_filtered = df.drop(df[df['ratingScore'] < 3].index)
print(df_filtered.head())  # Display DataFrame after dropping the rows
```

```python
# Access a row by index and a specific column (e.g., row 0, 'reviewTitle')
print(df.loc[0, 'reviewTitle'])

# Access multiple rows and columns using labels
print(df.loc[0:2, ['country', 'ratingScore', 'reviewTitle']])

# Access a row by position (e.g., row 1, column 3)
print(df.iloc[1, 3])

# Access a slice of rows and columns using positions
print(df.iloc[0:3, [0, 1, 2]])  # Access the first three rows and the first three columns
```

```python
# Filtering based on a condition (e.g., reviews with ratingScore > 4)
high_rating_reviews = df[df['ratingScore'] > 4]
print(high_rating_reviews.head())

# Filtering with multiple conditions (e.g., reviews where ratingScore > 4 and country is 'India')
filtered_reviews = df[(df['ratingScore'] > 4) & (df['country'] == 'India')]
print(filtered_reviews.head())

# Filtering using OR condition (e.g., ratingScore > 4 or country is 'India')
filtered_reviews_or = df[(df['ratingScore'] > 4) | (df['country'] == 'India')]
print(filtered_reviews_or.head())
```

```python
# Grouping data by 'country' and aggregating (e.g., counting reviews by country)
grouped_country = df.groupby('country').size()
print(grouped_country)

# Grouping by 'ratingScore' and calculating the mean of 'ratingScore' by country
grouped_mean_rating = df.groupby('country')['ratingScore'].mean()
print(grouped_mean_rating)

# Grouping by multiple columns (e.g., 'country' and 'ratingScore') and summing or counting
grouped_multi = df.groupby(['country', 'ratingScore']).size()
print(grouped_multi)
```

Program to work with visualization using seaborn

e. Create Hex Plot for Energy vs Serve Size

h. Create Boxplot of Energy per Category

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/India_Menu.csv'  # Update the file path
df = pd.read_csv(file_path)

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# a. Print different attributes of dataframe
print("Shape of the dataframe:", df.shape)
print("\nData Types:\n", df.dtypes)
print("\nFirst 5 Records:\n", df.head())
```

```python
# b. Create a Histogram of Energy (kCal)
plt.figure(figsize=(5, 4))
sns.histplot(df['Energy (kCal)'], kde=True, color='blue')
plt.title('Histogram of Energy (kCal)')
plt.xlabel('Energy (kCal)')
plt.ylabel('Frequency')
plt.show()
```

```python
# c. Create Scatter plot for Energy vs Sugars across the Items
plt.figure(figsize=(5, 4))
sns.scatterplot(x='Energy (kCal)', y='Total Sugars (g)', data=df, color='green')
plt.title('Scatter plot: Energy (kCal) vs Total Sugars (g)')
plt.xlabel('Energy (kCal)')
plt.ylabel('Total Sugars (g)')
plt.show()
```

```python
# d. Create Regression Plot for Energy vs Carbohydrates
plt.figure(figsize=(5, 4))
sns.regplot(x='Energy (kCal)', y='Total carbohydrate (g)', data=df, color='purple', scatter_kws={'s':20})
plt.title('Regression Plot: Energy (kCal) vs Total Carbohydrates (g)')
plt.xlabel('Energy (kCal)')
plt.ylabel('Total Carbohydrates (g)')
plt.show()
```

```python
# e. Create Hex Plot for Energy vs Serve Size
plt.figure(figsize=(8, 6))

# Convert 'Per Serve Size' to numeric, removing non-numeric characters
df['Per Serve Size'] = pd.to_numeric(df['Per Serve Size'].str.replace('[^0-9.]', '', regex=True), errors='coerce')

# Now create the hexbin plot
plt.hexbin(x=df['Energy (kCal)'], y=df['Per Serve Size'], gridsize=30, cmap='Blues')
plt.title('Hex Plot: Energy (kCal) vs Per Serve Size')
plt.xlabel('Energy (kCal)')
plt.ylabel('Per Serve Size')
plt.colorbar(label='Counts')  # Adds color bar to indicate counts
plt.show()
```

```python
# f. Create Pair Plot using Energy, Total fat, Carbohydrates, Sugars, Protein as features of dataset
plt.figure(figsize=(8, 6))
sns.pairplot(df[['Energy (kCal)', 'Total fat (g)', 'Total carbohydrate (g)', 'Total Sugars (g)', 'Protein (g)']])
plt.show()
```

```python
# g. Create multivariate scatter plot for Energy Variation with Sugars for each category
plt.figure(figsize=(8, 4))
sns.scatterplot(x='Total Sugars (g)', y='Energy (kCal)', hue='Menu Category', data=df)
plt.title('Multivariate Scatter Plot: Energy (kCal) vs Sugars by Category')
plt.xlabel('Total Sugars (g)')
plt.ylabel('Energy (kCal)')
plt.legend(loc='upper right')
plt.show()
```
